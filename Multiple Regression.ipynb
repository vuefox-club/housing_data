```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
from sklearn.datasets import load_boston
```

```python
boston_data=load_boston()
```

```python
df=pd.DataFrame(boston_data.data,columns=boston_data.feature_names)
```

```python
df.head()
```

```python
df.shape
```

```python
X=df
```

```python
y=boston_data.target
```

# Statsmodels

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
X_constant=sm.add_constant(X)
```

```python
pd.DataFrame(X_constant)
```

```python
df_test=pd.DataFrame(X_constant)
```

```python
sm.OLS?
```

```python
model=sm.OLS(y,X_constant)
```

```python
lr=model.fit()
```

```python
lr.summary()
```

# statsmodels.formula.api

```python
form_lr=smf.ols(formula='y~CRIM+ZN+INDUS+CHAS+NOX+RM+AGE+DIS+RAD+TAX+PTRATIO+B+LSTAT',data=df)
mlr=form_lr.fit()
```

```python
mlr.summary()
```

```python
form_test=smf.ols(formula='y~const + CRIM + ZN + INDUS + CHAS + NOX + RM',data=df_test)
mlr_test=form_test.fit()
```

```python
mlr_test.summary(0)
```

```python
form_lrtest=smf.ols('y~CRIM+ZN+CHAS+NOX',data=df)
model_test2=form_lrtest.fit()
model_test2.summary()
```

# Detecting Collinearity with Eigenvectors

```python
eigenvalues, eigenvectors = np.linalg.eig(df.corr())
```

```python
pd.options.display.float_format='{:,.4f}'.format
pd.Series(eigenvalues).sort_values()
```

```python
np.abs(pd.Series(eigenvectors[:,8])).sort_values(ascending=False)
```

# Standardise Variable to Identify Key Features()

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
result = pd.DataFrame(list(zip(model.coef_, df.columns)), columns = ['coefficient' , 'name']).set_index('name')
np.abs(result).sort_values(by='coefficient',ascending=False)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
scaler=StandardScaler()
standard_coefficient_linear_reg = make_pipeline(scaler, model)
```

```python
standard_coefficient_linear_reg.fit(X,y)
result=pd.DataFrame(list(zip(standard_coefficient_linear_reg.steps[1][1].coef_, df.columns)),
                   columns=['coefficient', 'name']).set_index('name')
np.abs(result).sort_values(by='coefficient', ascending=False)
```

# Use R2 to identify key features

```python
from sklearn.metrics import r2_score
```

```python
linear_reg=smf.ols('y~CRIM+ZN+INDUS+CHAS+NOX+RM+AGE+DIS+RAD+TAX+PTRATIO+B+LSTAT',data=df)
benchmark = linear_reg.fit()
r2_score(y, benchmark.predict(df))
```

```python
linear_reg=smf.ols('y~CRIM+ZN+INDUS+CHAS+NOX+RM+AGE+DIS+RAD+TAX+PTRATIO+B',data=df)
lr_without_Lstat = linear_reg.fit()
r2_score(y, lr_without_Lstat.predict(df))
```

```python
linear_reg=smf.ols('y~CRIM+ZN+INDUS+CHAS+NOX+RM+DIS+RAD+TAX+PTRATIO+B+LSTAT',data=df)
lr_without_AGE = linear_reg.fit()
r2_score(y, lr_without_AGE.predict(df))
```
